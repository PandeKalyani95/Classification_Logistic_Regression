## Creating dummies for categorical columns

```python
## create dummies for categorical vales

new_data = pd.get_dummies(data)
new_data.head()
```

## Scaling data in between 0 to 1 for better result

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
scaler = MinMaxScaler()
```

```python
# column 0 to 9 need to scale

var = new_data.iloc[:,:10].columns
var = var.tolist()
type(var)
```

```python
new_data[var] = scaler.fit_transform(new_data[var])
new_data.head()
```

```python
new_data.max()
```

## spliting data into trining and testing dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x = new_data.drop(columns=['y'])
Y = new_data['y']
```

```python
x_train, x_test, Y_train, Y_test= train_test_split(x,Y,test_size=0.2,random_state=2)
```

```python
print(x_train.shape)
print(x_test.shape)
```

## createing Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
logit = LogisticRegression()
```

## fit a model on traing dataset

```python
lr = logit.fit(x_train, Y_train)
```

## predict model by using test dataset

```python
Y_pred_test = lr.predict(x_test)
```

## checking Classifier evaluation

- score:- y_pred == y_obs
- accuracy score:- (score and accuracy did the same thing checking model accuracy.
- confusion metrics:- used to describe the performance of a classification model
- roc and roc_auc curve:- shows how good model is by graphical representation.
- classification report:- accuracy, precision, recall, F-1 score

```python
# score:
lr.score(x_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print('Confusion Matrix: ')

confusion_matrix(Y_test,Y_pred_test)
```

```python
print('Classification Report:-')

classification_report(Y_test,Y_pred_test)
```

```python
print('Accuracy Score:-')

accuracy_score(Y_test, Y_pred_test)
```
